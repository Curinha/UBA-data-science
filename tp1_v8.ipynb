```python
import pandas as pd
df = pd.read_csv('/content/review.csv')
df
```

```python
df = df[["review_id", "date"]]
df
```

```python
df = df.sort_values("date", ascending=True)
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df["date"] = pd.to_datetime(df["date"])
df
```

```python
df.describe()
```

```python
differences = []
df.reset_index(inplace=True)

for i in range (0, len(df)):
  if (i == 0):
    differences.append("00:00:00")
  else:
    differences.append(df["date"][i] - df["date"][i-1])

df["differences"] = differences
df
```

```python
df["differences"] = df["differences"].map(lambda x: x.seconds)
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(dpi=150)
sns.boxplot(x=df["differences"], color="#A1CF45")
plt.xlim(0,200)
plt.title("Distribución del tiempo entre reviews en el sitio")
plt.xlabel("Tiempo (s)")
```
