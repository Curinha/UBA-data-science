```python
df = df[["review_id", "stars", "text"]]
df
```

**StopWords**

No uso Stemmer ni Lemmatizer porque no considero necesario segun consigna. 
De hecho, creo que la idea es ver las 500 palabras más frecuentes sin estos normalizadores.

```python
from nltk.corpus import stopwords
from nltk import word_tokenize
import nltk
# nltk.download('punkt')
# nltk.download('wordnet')
# nltk.download('stopwords')
```

```python
stopwordsEN = set(stopwords.words('english'))
# stopwordsEN
```

```python
#Defino funcion para luego aplicarla en la columna

def removeStopWords(texto):
  textoSinStopwords = ""
  for palabra in word_tokenize(texto):
    if palabra not in stopwordsEN:
        textoSinStopwords = textoSinStopwords + " " + palabra
        
  return textoSinStopwords
```

```python
#Saco StopWords
df["text"] = df["text"].apply(removeStopWords)
df
```

```python
import random
random.choice(df["text"])
```

```python
textos = df["text"].tolist()
```

```python
#Solo me interesa ver si aparece o no. No la frecuencia. 
from sklearn.feature_extraction.text import CountVectorizer

vector = CountVectorizer(lowercase=True, max_features=500)
matrizVectores = vector.fit_transform(textos)
palabras = vector.get_feature_names_out()
print(palabras.size)
# print(palabras)
```

```python
X = matrizVectores.toarray()
X.shape
```

```python
df_nlp = pd.DataFrame(X, columns=palabras)
df_nlp
```

```python
stars = 0
news = 0
stars_4_word = []

#Agarro una palabra y la analizo por todas las noticias
for i in range (0,500):
  for j in df_nlp.index:
    if (df_nlp[palabras[i]][j] > 0):
      stars = stars + df["stars"][j]
      news = news + 1
  #termino la palabra en cuestion y hago el promedio
  stars_4_word.append(stars/news)
  stars = 0
  news = 0
```

```python
df_word_stars = pd.DataFrame(stars_4_word, index=palabras, columns=["stars"])
df_word_stars
```

```python
def predictor(texto):

  score = 0
  count = 0

  for palabra in word_tokenize(texto):
    if palabra in palabras:
        stars = df_word_stars.loc[palabra]
        score = score + stars
        count = count + 1
    #luego de buscar todas las palabras de la frase

  return (score/count)
```

```python
predictor("10 or 11 years and young")
```
