```python
#Transformo str a lista para iterar sobre los amigos. ACA ES DONDE TENGO EL PROBLEMA DE MEMORIA
df["friends"] = df["friends"].map(lambda x: x.split(", "))
df
```

```python
mean_list = []

for i in df.index:
    suma = 0
    count = 0 
    for friend in df["friends"][i]:
      try: 
        suma = suma + df.loc[str(friend), "friends_count"]
        count = count + 1
      except:
          continue

    if (count != 0): #si tiene amigos, agregamos el promedio a la lista.
      mean_list.append(suma/count)
    else: #si no tiene amigos, agregamos NaN a la lista
      mean_list.append(None)
```

```python
#Coincide con tamaño df
len(mean_list)
```

```python
df["mean_friends"] = mean_list
```

```python
df
```

```python
df["paradox"] = df["mean_friends"] > df["friends_count"]
df

#Para los casos borde donde no tienen amigos o no tenemos el registro de los amigos de los amigos, paradox será False.
#Pq a priori, no sabemos si se cumple.
```

```python
result = df["paradox"].sum() / len(df) 
print(result)

""" Solo usando 10000 registros de los 1987897, la paradoja se cumple un 75% de las veces"""
```
