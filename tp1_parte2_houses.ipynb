# VISUALIZACION 1 - OK

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("/content/data/train.csv")
df = df[["TotRmsAbvGrd","SalePrice"]]
df
```

```python
colors = ["#86bbd8","#00a5cf", "#33658a", "#3c7a89", "#f4d35e","#f6ae2d", "#f78154",  "#f26419", "#e5383b", "#b10f2e",  "#570000", "#372554" ]


plt.figure(dpi=200)
sns.barplot(y='SalePrice', x="TotRmsAbvGrd", data=df, palette=sns.color_palette(colors, 12), ci=None)
plt.xlabel("Habitaciones (u)")
plt.ylabel("Precio de venta de la propiedad (USD)")
plt.title("Distribución del promedio del precio de venta de una propiedad \n en base al numero de habitaciones (sin contemplar baños)")
```

# VISUALIZACION 2 - OK

```python
df = pd.read_csv("/content/data/train.csv")
df = df[["OverallQual","SalePrice"]]
df
```

```python
plt.figure(figsize=(8, 6), dpi=170)

colors = ["#86bbd8", "#33658a", "#3c7a89", "#f4d35e","#f6ae2d", "#f78154",  "#e5383b", "#b10f2e","#570000", "#372554" ]

sns.boxplot(y='SalePrice', x="OverallQual", data=df, palette=sns.color_palette(colors, 10));
plt.xlabel("Material general y calidad de acabado (1-10)")
plt.ylabel("Precio de venta de la propiedad (USD)")
plt.title("Distribución del precio de venta de una propiedad \n en base al indice de calidad general")
```
